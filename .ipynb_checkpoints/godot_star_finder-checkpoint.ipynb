```python
import requests
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import re
```

```python
def scrape_assets(max_pages=1):
    all_assets = []
    
    for page in range(max_pages):
        url = f"https://godotengine.org/asset-library/asset?max_results=500&page={page}&sort=name"
        print(f"Scraping page {page}...")
        
        try:
            response = requests.get(url)
            response.raise_for_status()
            
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Find all asset items
            asset_items = soup.find_all('li', class_='asset-item')
            print(f"Found {len(asset_items)} assets on page {page}")
            
            for item in asset_items:
                asset_data = {}
                
                # Get title
                title_elem = item.find('h4')
                if title_elem:
                    asset_data['title'] = title_elem.get_text(strip=True)
                
                # Get asset ID from link
                link_elem = item.find('a', class_='asset-header')
                if link_elem and link_elem.get('href'):
                    href = link_elem['href']
                    asset_id = href.split('/')[-1]
                    asset_data['asset_id'] = asset_id
                    asset_data['asset_url'] = f"https://godotengine.org{href}"
                
                # Get icon URL (often points to GitHub)
                img_elem = item.find('img', class_='media-object')
                if img_elem and img_elem.get('src'):
                    icon_src = img_elem['src']
                    asset_data['icon_url'] = icon_src
                    
                    # Extract GitHub repo from icon URL
                    if 'github.com' in icon_src or 'githubusercontent.com' in icon_src:
                        github_match = re.search(r'github\.com/([^/]+/[^/]+)', icon_src.replace('githubusercontent.com', 'github.com'))
                        if github_match:
                            asset_data['github_repo'] = github_match.group(1)
                            asset_data['github_url'] = f"https://github.com/{github_match.group(1)}"
                
                # Get tags/categories
                tags = []
                tag_spans = item.find_all('span', class_='label')
                for span in tag_spans:
                    tag_text = span.get_text(strip=True)
                    tags.append(tag_text)
                asset_data['tags'] = tags
                
                # Get category (first primary tag)
                primary_tag = item.find('span', class_='label-primary')
                if primary_tag:
                    asset_data['category'] = primary_tag.get_text(strip=True)
                
                # Get Godot version
                version_tag = item.find('span', class_='label-info')
                if version_tag:
                    asset_data['godot_version'] = version_tag.get_text(strip=True)
                
                # Get license
                license_tag = item.find('span', class_='label-default')
                if license_tag:
                    asset_data['license'] = license_tag.get_text(strip=True)
                
                # Get author
                footer = item.find('div', class_='asset-footer')
                if footer:
                    author_link = footer.find('a')
                    if author_link:
                        author_text = author_link.get('title', '')
                        if 'Search assets by' in author_text:
                            author = author_text.replace('Search assets by', '').strip(" '\"")
                            asset_data['author'] = author
                
                if asset_data:  # Only add if we got some data
                    all_assets.append(asset_data)
            
            print(f"Extracted {len(all_assets)} total assets so far")
            
        except Exception as e:
            print(f"Error scraping page {page}: {e}")
        
        time.sleep(2)  # Be respectful
    
    return all_assets

print("Starting Godot Asset Library scraping...")
assets = scrape_assets(max_pages=20)

if assets:    
    # Save GitHub repos to CSV for later use
    github_assets = [a for a in assets if a.get('github_url')]
    print(f"\n=== SAVING {len(github_assets)} GITHUB REPOS ===")
    
    with open('godot_github_repos.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['title', 'github_url', 'category', 'license', 'author'])
        for asset in github_assets:
            writer.writerow([
                asset.get('title', ''),
                asset.get('github_url', ''),
                asset.get('category', ''),
                asset.get('license', ''),
                asset.get('author', '')
            ])
    
    print("Saved GitHub repos to 'godot_github_repos.csv'")
else:
    print("No assets found!")
```

```python
def get_github_stats(github_url):
    """Extract stars and forks from a GitHub repository page"""
    try:
        response = requests.get(github_url, timeout=10)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Look for star count - GitHub uses different selectors
        stars = 0
        forks = 0
        
        # Try different selectors for stars
        star_selectors = [
            '#repo-stars-counter-star',
            'a[href$="/stargazers"] strong',
            'a[href$="/stargazers"]',
            '[data-target="repository-statistics.stars"]'
        ]
        
        for selector in star_selectors:
            star_elem = soup.select_one(selector)
            if star_elem:
                star_text = star_elem.get_text(strip=True)
                # Extract number from text like "1.2k" or "123"
                star_match = re.search(r'([\d.]+)([km]?)', star_text.lower())
                if star_match:
                    num = float(star_match.group(1))
                    suffix = star_match.group(2)
                    if suffix == 'k':
                        num *= 1000
                    elif suffix == 'm':
                        num *= 1000000
                    stars = int(num)
                    break
        
        # Try different selectors for forks
        fork_selectors = [
            '#repo-network-counter',
            'a[href$="/forks"] strong',
            'a[href$="/network/members"]',
            '[data-target="repository-statistics.forks"]'
        ]
        
        for selector in fork_selectors:
            fork_elem = soup.select_one(selector)
            if fork_elem:
                fork_text = fork_elem.get_text(strip=True)
                fork_match = re.search(r'([\d.]+)([km]?)', fork_text.lower())
                if fork_match:
                    num = float(fork_match.group(1))
                    suffix = fork_match.group(2)
                    if suffix == 'k':
                        num *= 1000
                    elif suffix == 'm':
                        num *= 1000000
                    forks = int(num)
                    break
        
        return stars, forks
        
    except Exception as e:
        print(f"Error fetching {github_url}: {e}")
        return 0, 0

def enrich_with_github_stats(csv_file='godot_github_repos.csv'):
    """Add stars and forks columns to the existing CSV"""
    
    # Read existing CSV
    repos = []
    with open(csv_file, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            repos.append(row)
    
    print(f"Loaded {len(repos)} repositories from {csv_file}")
    
    # Add stars and forks to each repo
    for repo in repos:
        repo['stars'] = 0
        repo['forks'] = 0
    
    print("Fetching GitHub stats...")
    
    for i, repo in enumerate(repos):
        github_url = repo['github_url']
        print(f"Processing {i+1}/{len(repos)}: {github_url}")
        
        stars, forks = get_github_stats(github_url)
        repo['stars'] = stars
        repo['forks'] = forks
        
        print(f"  Stars: {stars}, Forks: {forks}")
        
        # Be respectful - longer delay for GitHub
        time.sleep(3)
        
        # Save progress every 50 items
        if (i + 1) % 50 == 0:
            print(f"Saving progress... ({i+1}/{len(repos)})")
            save_repos_to_csv(repos, 'godot_github_repos_with_stats.csv')
    
    # Final save
    output_file = 'godot_github_repos_with_stats.csv'
    save_repos_to_csv(repos, output_file)
    print(f"\nSaved complete data to {output_file}")
    
    return repos

def save_repos_to_csv(repos, filename):
    """Save repos list to CSV file"""
    if not repos:
        return
        
    fieldnames = repos[0].keys()
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(repos)
```
